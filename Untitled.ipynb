```python
import pandas as pd
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster, cophenet
from scipy.spatial.distance import pdist
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
import numpy as np
```

```python
data = pd.read_csv("temper.csv")
```

# CAH

# Dendrogramme obtenu par la CAH

```python
# Extract city names and temperature data, excluding Latitude and Longitude columns
city_names = data.iloc[:, 0]  # Assuming the first column contains city names
temperature_data = data.iloc[:, 1:-2]  # Assuming the temperature data is from the second column to the second-to-last column
```

```python
# Perform hierarchical/agglomerative clustering
linked = linkage(temperature_data, 'ward')  # Perform clustering without transposing temperature data
```

```python
# Plotting the dendrogram
plt.figure(figsize=(12, 8))
dendrogram(linked,
           orientation='top',
           labels=city_names.values,  # Use .values to convert city_names to a numpy array
           distance_sort='descending',
           show_leaf_counts=True)
plt.title('Dendrogram of Monthly Average Temperatures for French Cities')
plt.xlabel('Cities')
plt.ylabel('Euclidean Distances')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# Evolution de l‚Äôinertie en fonction du nombre de groupes

```python
# Ensure the temperature data is in numeric format
temperature_data = temperature_data.apply(pd.to_numeric, errors='coerce')

# Drop rows with NaN values, if any
temperature_data.dropna(inplace=True)

# Calculate inter-cluster distances, intra-cluster distances, and total inertia gain for different numbers of clusters
max_clusters = 15  # Maximum number of clusters
inter_cluster_distances = []
intra_cluster_distances = []
total_inertia_gains = []

for n_clusters in range(1, max_clusters + 1):
    # Perform hierarchical/agglomerative clustering
    linked = linkage(temperature_data, 'ward')
    inertia = linked[-1, 2] if n_clusters > 1 else 0
    total_inertia_gains.append(inertia)

    # Calculate inter-cluster distance
    inter_cluster_dist = linked[-(n_clusters - 1), 2]
    inter_cluster_distances.append(inter_cluster_dist)

    # Calculate cluster labels for the final configuration
    cluster_labels = fcluster(linked, n_clusters, criterion='maxclust')

    # Calculate intra-cluster distance for the final configuration
    intra_cluster_dist = 0
    for cluster_num in range(1, n_clusters + 1):
        cluster_points = temperature_data[cluster_labels == cluster_num]
        centroid = cluster_points.mean(axis=0)
        intra_cluster_dist += np.sum(np.linalg.norm(cluster_points - centroid, axis=1))
    intra_cluster_distances.append(intra_cluster_dist)

# Plotting inter-cluster, intra-cluster distances, and total inertia gain against the number of clusters
plt.figure(figsize=(8, 6))
plt.plot(range(1, max_clusters + 1), inter_cluster_distances, marker='o', label='I-inter')
plt.plot(range(1, max_clusters + 1), intra_cluster_distances, marker='o', label='I-intra')
plt.plot(range(1, max_clusters + 1), total_inertia_gains, marker='o', label='IG')
plt.title('Evolution of Distances and IG vs Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Distance/Inertia')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Evolution du critere R^2(k)

```python
# Calculate inter-cluster distances and total inertia gain for different numbers of clusters
max_clusters = 15  # Maximum number of clusters
inter_cluster_distances = []
total_inertia_gains = []

for n_clusters in range(1, max_clusters + 1):
    # Perform hierarchical/agglomerative clustering
    linked = linkage(temperature_data, 'ward')
    inertia = linked[-1, 2] if n_clusters > 1 else 0
    total_inertia_gains.append(inertia)

    # Calculate inter-cluster distance
    inter_cluster_dist = linked[-(n_clusters - 1), 2]
    inter_cluster_distances.append(inter_cluster_dist)

# Calculate R-squared-like metric values
r_squared_values = [(total_inertia_gains[i] - inter_cluster_distances[i]) / total_inertia_gains[i] for i in range(max_clusters)]

# Plotting R-squared-like values against the number of clusters
plt.figure(figsize=(8, 6))
plt.plot(range(1, max_clusters + 1), r_squared_values, marker='o')
plt.title('Evolution of R-squared(k)')
plt.xlabel('Number of Clusters')
plt.ylabel('R-squared(k)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Representation des clusters

```python
from sklearn.decomposition import PCA
latitude = data['Latitude']
longitude = data['Longitude']
city_names = data['Unnamed: 0']  # Assuming the column name holds city names

# Perform hierarchical/agglomerative clustering to get cluster labels
num_clusters = 3  # Choose the number of clusters
linked = linkage(temperature_data, 'ward')
cluster_labels = fcluster(linked, num_clusters, criterion='maxclust')

# Reduce data dimensionality to 2 using PCA
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(temperature_data)

# Plot clusters in a scatter plot with reversed axes
plt.figure(figsize=(8, 6))
for i in range(1, num_clusters + 1):
    plt.scatter(longitude[cluster_labels == i], latitude[cluster_labels == i], label=f'Cluster {i}')

# Annotate each point with city name
for i, city in enumerate(city_names):
    plt.text(longitude[i], latitude[i], city, fontsize=8, ha='left', va='bottom')

plt.title('Representation of Clusters based on Longitude and Latitude')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# KMeans

```python
Data = pd.read_csv("temper.csv")
Data.reset_index(inplace=True,drop=True)
Data.rename(columns={'Unnamed: 0': 'City'}, inplace=True)
location_data = Data.loc[:,['City','Latitude','Longitude']]
```

# Classes obtenues par la methode des k-moyennes avec k = 3

```python
from sklearn.cluster import KMeans,BisectingKMeans
# Perform K-means clustering with k=3
kmeans = BisectingKMeans(n_clusters=3)
kmeans.fit(location_data.iloc[:, 1:])
cluster_labels = kmeans.labels_
# Calculate inter-cluster distance and total inertia gain
inter_cluster_distances = kmeans.transform(location_data.iloc[:, 1:])[range(len(location_data.iloc[:, 1:])), cluster_labels]
inertia = kmeans.inertia_
print(cluster_labels,"\n",inertia,"\n",inter_cluster_distances)

# Calculate the ratio of I-inter to IG for each cluster
unique_ratios = inter_cluster_distances / inertia
# Plot scatter plots for each class with specified ratios
plt.figure(figsize=(12, 4 * len(unique_ratios)))
for i, ratio in enumerate(unique_ratios):
    plt.subplot(len(unique_ratios), 1, i + 1)
    plt.title(f'Class {i+1} (I-inter/IG = {ratio * 100:.1f}%)')

    # Plot each cluster separately for the specific ratio and annotate points with city names
    for idx in range(3):
        cluster_points = location_data[cluster_labels == idx]
        plt.scatter(cluster_points.iloc[:, 0], cluster_points.iloc[:, 1], label=f'Cluster {idx}')
        for city_index, (_, row) in enumerate(cluster_points.iterrows()):
            plt.text(row[0], row[1], data.iloc[city_index, 0], fontsize=8)  # Annotate with city names

    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.legend()
    plt.tight_layout()

plt.show()
```
